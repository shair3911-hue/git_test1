# Indexing and slicing of sequences

```python
name = "Alice"
print(name)
```

```python
len(name)
```

```python
name[0]
```

```python
name[-1]
```

```python
name[1:4] #it means from index 1 to index 3 (4 is not included)
```

```python
name[0:] #it means from index 0 to the end of the string
```

```python
name[2:] #it means from index 2 to the end of the string
```

```python
name[-1:-5] #it means from index -1 to index -4 (-5 is not included) but it will return an empty string because the start index is greater than the end index
name[-5:-1] #it means from index -5 to index -2 (-1 is not included) but it will return an empty string because the start index is less than the end index
```
