```python
import matplotlib.pyplot as plt

# Create dummy data
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 5, 7]

# Create line plot
plt.figure(figsize=(8, 6))
plt.plot(x, y, marker='o', label='Data')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line Plot from Dummy Data')
plt.legend()
plt.grid(True)
plt.show()
```
