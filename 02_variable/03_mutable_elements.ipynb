# Mutable Elements: Lists, Sets, and Dictionaries

```python
list=["apple", "banana", "orange", "grape", "kiwi"]
print(list)
```

```python
list[3]="melon" #Kisi variable se kisi value ko change krna he to to us value ka index lga kr us value k equall kr do jo add krna he
```

```python
print(list)
```

```python
tuple=("apple", "banana", "orange", "grape", "kiwi")
print(tuple)
```

```python
tuple[3]="melon" #Tuple is immutable so it will give the error because we cannot change the value of tuple by using index
print(tuple) 
```

```python
t = 1, 2, 3   # still a tuple
print(t)
```

```python
t[2]=4
print(t) #This will give an error because tuple is immutable and we cannot change its value by using index.
```

## so remember: List is mutable and written with [], while tuple is immutable and written with ()
